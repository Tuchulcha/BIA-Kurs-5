```python
import pandas as pd
import numpy as np
from tqdm.notebook import  tqdm #Om du kör i vscode, använd inte .notebook
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
## Helper-funktioner
```

```python
#Genererar datum
def generate_random_date(start_date, range_in_days):
    days_to_add = np.random.randint(range_in_days)
    random_date = np.datetime64(start_date) + days_to_add
    return random_date

generate_random_date("2015-01-01", 20)
```

```python
def hr_fmt(x, pos = 0):
    if x >= 1_000_000_000:
        return f'{x*1e-9:.0f}B'
    elif x >= 1_000_000:
        return f'{x*1e-6:.0f}M'
    elif x >= 1_000:
        return f'{x*1e-3:.0f}k'
    else:
        return f'{x:.0f}'


    
# Create a FuncFormatter object based on the custom function
formatter = FuncFormatter(hr_fmt)
    
def format_axes(plt):
    # Applying the formatter to the y-axis
    ax = plt.gca()  # Get the current axes
    ax.yaxis.set_major_formatter(formatter)
```

```python
attributes = ['name', 'price', 'weight', 'brand', 'cost']

products = [
    ['ifån', 7900, 4.2, 'Apple', 7900 * 0.7],
    ['MacBrick', 15900, 3.2, 'Apple', 15900 * 0.7],
    ['Papple PC', 14500, 4.2, 'Apple', 14500 * 0.7],
    ['Japple Watch', 3900, 0.2, 'Apple', 3900 * 0.7],
    ['FiPad', 5500, 1.2, 'Apple', 5500 * 0.7],

    ['Hony HD Webcam', 1500, 3.2, 'Sony', 1500*0.8],
    ['Rony Sound Freak', 1200, 0.2, 'Sony', 1200*0.8],
    ['Cony SD Webcam', 500, 3.2, 'Sony', 500*0.8],
    ['Flony Bravia TV', 540, 5.2, 'Sony', 540*0.8],

    ['Tamsung Crystal TV', 540, 5.2, 'Samsung', 540*0.8],
]

df_products = pd.DataFrame(products, columns=attributes)
```

```python
linear_values = np.arange(len(products))

exp_values = np.exp(linear_values / 2)

probabilities = exp_values / exp_values.sum()

def random_product():
    index =  np.random.choice(len(products), p = probabilities)
    return products[index]
```

```python
## Generate Random Sales
def generate_sales_no_batch(sales_to_generate):
    
    sales = []
    dates = []
    
    #Generate Sales
    for i in tqdm(range(sales_to_generate), desc="Generating Sales"):
        product = random_product()
        date = generate_random_date("2015-01-01", 365*5)
        sales.append(product)
        dates.append(date)

    df_sales = pd.DataFrame(sales, columns=attributes)
    df_sales['date'] = dates
    
    return df_sales

#generate_sales_no_batch(10_000_000) #Understreck räknas inte i koden, gör det lättare att se för människor
```

```python
def generate_sales_batched(sales_to_generate, batch_size=1000):
    
    sales = []
    dates = []
    
    batch_size = 1000
    outer_range = int (sales_to_generate / 1000)
    
    #Generate Sales
    #for i in tqdm(range(sales_to_generate), desc="Generating Sales"): #Vi delar upp det istället!
    for i in tqdm(range(outer_range), desc="Generating Sales"):     #tqdm gör att det går att se progress, coolt!
        for j in range(batch_size):
            product = random_product()
            date = generate_random_date("2015-01-01", 365*5)
            sales.append(product)
            dates.append(date)

    df_sales = pd.DataFrame(sales, columns=attributes)
    df_sales['date'] = dates
    
    return df_sales

#generate_sales_batched(10_000_000) #Understreck räknas inte i koden, gör det lättare att se för människor
```

```python
##  Linus missnöjd med fördelningen över datum
```

```python
%%time
df_sales = generate_sales_batched(1000_000)
```

```python
df_sales.info()
```

```python
## Plot Data
### Sales per Year
```

```python
#extract  year  from  each  date
df_sales['year'] = df_sales['date'].dt.year #funkar på rätt datatyp (datetime)
df_sales.head()
```

```python
sales_per_year = df_sales['year'].value_counts().sort_index()
sales_per_year
```

```python
plt.figure(figsize=(10, 5)) #(bred,hög)
sns.barplot(x = sales_per_year.index, y = sales_per_year.values)

for i, value in enumerate(sales_per_year.values):
    plt.text(i, value, hr_fmt(value), ha='center', va='bottom' )

format_axes(plt) # Här används FuncFormatter funktion-i-funktionigt

plt.xlabel('Year')
plt.ylabel('Number of Sales')
plt.title('Number of Sales from 2015-2020')

plt.show()
```

```python
df_sales['profit'] = df_sales['price'] - df_sales['cost']

total_profit_per_product = df_sales.groupby('name')['profit'].sum().sort_values(ascending=False)

total_profit_per_product
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=total_profit_per_product.index, y=total_profit_per_product.values)

for i, value in enumerate(total_profit_per_product.values):
    plt.text(i, value, hr_fmt(value), ha='center', va = 'bottom' )

plt.xticks(rotation=45)
format_axes(plt)
plt.show()
```

```python
daily_sales = df_sales.groupby('date').size()

daily_sales
```

```python
rolling_avg = daily_sales.rolling(window=30).mean()
plt.figure(figsize=(15,5))

ax = plt.gca()
ax.set_ylim(450, 650)

daily_sales.plot(label="Daily Sales", alpha=0.4)
rolling_avg.plot(color='red', label="Daily Sales", linestyle="--", alpha=0.8)

plt.xlabel('Date')
plt.ylabel('Daily Sales')
plt.title('Daily Sales with 30 day rolling Average')
plt.show()
```
