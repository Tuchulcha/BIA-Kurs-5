# Uppgift: Generera data och visualisera

## Bakgrund

Visualisering är ett kraftfullt verktyg för att kommunicera data och information. Det kan hjälpa oss attupptäcka mönster och trender i datan och förstå vad den betyder. 
I denna uppgift kommer du att fåträna på att generera data och visualisera den.
## Exempel: Försäljning av produkter över tid

Skapa en visualisering som tydligt visar att försäljningen av produkter är högre under andra halvan avåret än under första halvan.
## Steg 1: Generera datan

Denna kod genererar en lista med försäljningsdata för varje dag under 2023. Du kan använda dennadata för att skapa en visualisering.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Number of days in each month
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Create an empty list to store sales data
sales_data = []

# Generate sales data for each month
for month, days in enumerate(days_in_month):
    if month < 6: # First half of the year
        sales = np.random.randint(70, 80, days).tolist()
    else: # Second half of the year
        sales = np.random.randint(110, 160, days).tolist()    
    sales_data.extend(sales)
    
# Create a dataframe with the sales data
df = pd.DataFrame({
    'Date': pd.date_range(start="2023-01-01", end="2023-12-31"),
    'Sales': sales_data
})

df.info()
```

## Steg 2: Visualisera data

Här kan du använda ett bibliotek som matplotlib, seaborn eller något annat visualiseringsverktyg:

```python
# Set a style
sns.set(style="whitegrid")
# Create a figure and axes
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x="Date", y="Sales")
# Set title and labels
plt.title("Försäljning av produkter under 2023")
plt.xlabel("Datum")
plt.ylabel("Antal sålda produkter")

days_mean = 30
df[f'{days_mean}_Day_Avg'] = df['Sales'].rolling(window = days_mean).mean()
sns.lineplot(data=df, x="Date", y=f'{days_mean}_Day_Avg', label=f"{days_mean}-dagars genomsnitt", color='red')

# Show the plot
plt.show()
```

```python
#days = pd.date_range(start="2023-01-01", end="2023-12-31")

days2 = pd.period_range(start="2023-01-01", periods=70).to_timestamp() #*

before_campain2 = np.linspace(1100,1500,30) + np.random.randint(-50, +50, 30)

during_campain2 = np.linspace(1500, 3000, 10) + np.random.randint(-50, +50, 10)

after_campain2 = np.linspace(3000,3250,30) + np.random.randint(-50, +50, 30)

users2 = np.concatenate([before_campain2, during_campain2, after_campain2])

df2 = pd.DataFrame({
    'Date' : days2,
    'Users': users2
})


# *Convert period to timestamp Men det har jag gjort upp
#df2['Date'] = df2['Date'].dt.to_timestamp()

sns.lineplot(data=df2, x= "Date", y="Users", label="Antal Användare")

# Start: Marknadsföringskampanj 01-31
plt.axvspan("2023-01-31", "2023-02-09", alpha = 0.1, color='green', label='Marknadsforing') #Ger det färgade området, alpha är opacitet
# Slut: 2023-02-09

plt.legend()
plt.title("ökning av användare")
plt.show()
```

```python
days = pd.period_range(start="2023-05-01", periods=30).to_timestamp()

rainfall = np.random.randint(0, 20, 30)

visitors = 1000 - rainfall * 30

plt.figure(figsize=(16,8))

df = pd.DataFrame({
    'Datum': days,
    'Nederbörd': rainfall,
    'Besökare': visitors
})

fig, ax1 = plt.subplots(figsize=(14,6))
ax1.set_xlabel('Datum')
ax1.set_ylabel('Nederbörd', color='Blue')
sns.lineplot(data=df, x='Datum', y='Nederbörd', ax=ax1, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Besökare', color= 'green')
sns.lineplot(data=df, x='Datum', y='Besökare', ax=ax2, color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.show()
```

```python
days = pd.period_range(start="2023-05-01", periods=30).to_timestamp()

rainfall = np.random.randint(0, 20, 30)

visitors = 1000 - rainfall * 20 + np.random.randint(-400, 50, 30)

df = pd.DataFrame({
    'Datum': days,
    'Nederbörd': rainfall,
    'Besökare': visitors
})

plt.figure(figsize=(10,6))

plt.scatter(x=rainfall, y=visitors)
sns.regplot(x=rainfall, y=visitors)

corr_rain_vis = (df['Nederbörd'].corr(df['Besökare']))
plt.title(f"Korrelation: {corr_rain_vis:.2f}")

plt.show()
```

3. Visa medarbetares produktivitet som avhängt på antal timmar sömn

```python
DataP = 200

#random.uniform(5, 10, 1000) Ger 1000 random decimaltal 5-10

sleep_hours = np.random.uniform(4, 12, DataP)

performance = (-20 * (sleep_hours-8)**2 +250 + np.random.normal(0, 10, DataP))

plt.figure(figsize=(10,6))
sns.regplot(x=sleep_hours, y=performance, order=2)

plt.title("Sömn & Produktivitet")
plt.xlabel("Sömn")
plt.ylabel("Produktivitet")

plt.show()
```
