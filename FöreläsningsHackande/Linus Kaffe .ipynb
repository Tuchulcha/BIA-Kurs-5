```python
import matplotlib.pyplot as plt
import numpy as np
coffee_intake = np.linspace(0,10,100)

plt.figure(figsize=(10,6))
plt.plot(coffee_intake, coffee_intake)
plt.grid(True)
plt.show()
```

```python
productivity = coffee_intake * 10
plt.figure(figsize=(10,6))
plt.plot(coffee_intake, productivity)
plt.grid(True)
plt.show()
```

```python
def coffee_plot(intake, productivity):
    plt.figure(figsize=(10,6))
    plt.plot(coffee_intake, productivity)
    plt.title("Coffee intake and productivity")
    plt.xlabel("Cups of coffee")
    plt.ylabel("Emails per hour")
    plt.grid(True)
    plt.show()
```

```python
coffee_plot(coffee_intake, productivity)
```
