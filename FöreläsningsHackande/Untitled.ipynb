```python
#Genererar datum
def generate_random_date(start_date, range_in_days):
    days_to_add = np.random.randint(range_in_days)
    random_date = np.datetime64(start_date) + days_to_add
    return random_date

generate_random_date("2015-01-01", 20)
```

```python
attributes = ['name', 'price', 'weight', 'brand', 'cost']

products = [
    ['ifån', 7900, 4.2, 'Apple', 7900 * 0.7],
    ['MacBrick', 15900, 3.2, 'Apple', 15900 * 0.7],
    ['Papple PC', 14500, 4.2, 'Apple', 14500 * 0.7],
    ['Japple Watch', 3900, 0.2, 'Apple', 3900 * 0.7],
    ['FiPad', 5500, 1.2, 'Apple', 5500 * 0.7],

    ['Hony HD Webcam', 1500, 3.2, 'Sony', 1500*0.8],
    ['Rony Sound Freak', 1200, 0.2, 'Sony', 1200*0.8],
    ['Cony SD Webcam', 500, 3.2, 'Sony', 500*0.8],
    ['Flony Bravia TV', 540, 5.2, 'Sony', 540*0.8],

    ['Tamsung Crystal TV', 540, 5.2, 'Samsung', 540*0.8],
]

df_products = pd.DataFrame(products, columns=attributes)
```

```python
linear_values = np.arange(len(products))

exp_values = np.exp(linear_values / 2)

probabilities = exp_values / exp_values.sum()

def random_product():
    index =  np.random.choice(len(products), p = probabilities)
    return products[index]
```

```python
## Generate Random Sales
def generate_sales_no_batch(sales_to_generate):
    
    sales = []
    dates = []
    
    #Generate Sales
    for i in tqdm(range(sales_to_generate), desc="Generating Sales"): #Vi delar upp det istället!
        product = random_product()
        date = generate_random_date("2015-01-01", 365*5)
        sales.append(product)
        dates.append(date)

    df_sales = pd.DataFrame(sales, columns=attributes)
    df_sales['date'] = dates
    
    return df_sales

generate_sales_no_batch(10_000_000) #Understreck räknas inte i koden, gör det lättare att se för människor
```
