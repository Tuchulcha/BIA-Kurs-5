```python
# Python code to generate DAX SWITCH function
dax_code = "Switch_LPPI = \nSWITCH(\n    TRUE(),"

# Loop through 1 to 100 to generate the lines
for i in range(1, 101):
    dax_code += f"\n    [LPPI_W_parameter Value] = {i}, {i * 0.01:.2f},"

# Remove the trailing comma and close the SWITCH function
dax_code = dax_code.rstrip(",") + "\n)"

# Print or save the generated DAX code
print(dax_code)
```
